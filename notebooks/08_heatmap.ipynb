```python
import itertools
import numpy as np
import ujson as json
import pandas as pd
import networkx as nx
import holoviews as hv
import seaborn as sns
import matplotlib.pyplot as plt
from utils import load_suppliers

hv.extension('bokeh')

%matplotlib inline
```

```python
df_contracts = pd.read_parquet('../data/contrataciones_network.parquet', engine='pyarrow')
df_unidades = df_contracts.loc[:, ['unidad_id', 'unidad_nombre']].drop_duplicates()
# Only tender_status = 'complete' and currency 'MXN'
cond = (
    (df_contracts.tender_status == 'complete') 
    & (df_contracts.award_currency == 'MXN') 
    & (df_contracts.contract_currency == 'MXN')
)
df_contracts = df_contracts.loc[cond].drop(['award_currency', 'contract_currency', 'award_status'], axis=1)
df_contracts = df_contracts.sort_values('contract_period_start')
df_contracts = df_contracts.rename(columns={'award_supplier_name': 'supplier_name', 'award_supplier_id': 'supplier_id'})

# TODO: poner una columna solo con el orden temporal
print(df_contracts.shape)
df_contracts.head(2)
```

```python
df_participantes = (pd.read_parquet('../data/contrataciones_parties.parquet')
        .drop(['entidad_role', 'unidad_role', 'entidad_nombre', 'ID'], axis=1))
# nos quedamos con tenders completos
df_participantes = df_participantes.loc[df_participantes.tender_status == 'complete']
df_participantes = df_participantes.rename(columns={'integrante_id': 'supplier_id', 'integrante_name': 'supplier_name'})
print(df_participantes.shape)
df_participantes.head(2)
```

```python
# que el corte sea 100?
df_tenders = (df_participantes.groupby(['entidad_id', 'unidad_id']).tender_id.nunique()
              .reset_index().rename(columns={'tender_id': 'n_tenders'}))
print(df_tenders.shape)
df_tenders.head(3)
```

```python
df_tenderers = (df_participantes.groupby(['entidad_id', 'unidad_id']).supplier_id.nunique()
                                .reset_index().rename(columns={'supplier_id': 'unique_tenderers'}))
print(df_tenderers.shape)
df_tenderers.head(3)
```

```python
df_ganadores = df_participantes.loc[df_participantes.roles == 'supplier,tenderer']
df_ganadores = (df_ganadores.groupby(['entidad_id', 'unidad_id']).supplier_id.nunique()
                            .reset_index().rename(columns={'supplier_id': 'unique_winners'}))
print(df_ganadores.shape)
df_ganadores.head(3)
```

```python
df_aux = pd.merge(df_tenders, df_tenderers, on=['entidad_id', 'unidad_id'], how='inner')
df_aux = pd.merge(df_aux, df_ganadores, on=['entidad_id', 'unidad_id'], how='inner')
print(df_aux.shape)
df_aux.head(2)
```

```python
scatter = hv.Scatter(
    df_aux, kdims=['unique_tenderers',], 
    vdims=['unique_winners', 'n_tenders', 'unidad_id', 'entidad_id']
)
scatter = scatter.opts(
    height=600, width=800, tools=['hover'],
    logz=True,
    size=9, alpha=0.8, 
    color='n_tenders', cmap='viridis', colorbar=True
)
scatter
```

# Pairwise 2

```python
# e = 'SAGARPA-261'
# e = 'CONAGUA-94'
e = 'CFE-76'
# e = 'SEDESOL-263'
# u = 'SMA001013UH9-016B00001'
# u = 'CNA890116SF2-016B00985'  # raro anterior

df_mike = (df_participantes.loc[(df_participantes.entidad_id == e)])
df_mike = df_mike.groupby(['tender_id', 'supplier_id'], as_index=False).roles.count()
tenders_mas_de_dos = (df_mike.groupby('tender_id', as_index=False).supplier_id.count()
                             .query('supplier_id > 1').tender_id.unique())
df_mike = df_mike.loc[df_mike.tender_id.isin(tenders_mas_de_dos)].drop('roles', axis=1)
print(df_mike.shape)
df_mike.head(2)
```

```python
pairwise = []
for g, df_sub in df_mike.groupby('tender_id'):
    suppliers = sorted(df_sub.supplier_id.unique())
    combinaciones = itertools.combinations(suppliers, 2)
    for c in combinaciones:
        pairwise.append(
            {'tender_id': g, 'par': '|'.join(c)}
        )
df_pairwise = pd.DataFrame(pairwise)
print(df_pairwise.shape)
df_pairwise.head(3)
```

```python
# seleccionamos los más ganadores
df_contracts_e = df_contracts.loc[(df_contracts.entidad_id == e)]
df_contracts_e = df_contracts_e.loc[:, ['tender_id', 'supplier_id', 'contract_amount']].drop_duplicates()

df_contratos = (df_contracts_e.groupby('supplier_id', as_index=False).contract_amount.count()
                              .rename(columns={'contract_amount': 'n_contracts'}))

# Umbral
NC = 20
mas_ganadores = set(df_contratos.loc[df_contratos.n_contracts >= NC].supplier_id.unique())
print(len(mas_ganadores))

indices = []
for row in df_pairwise.par.str.split('|'):
    if any(r in mas_ganadores for r in row):
        indices.append(True)
    else:
        indices.append(False)
        
# Pairwise filtrado
df_pairwise = df_pairwise.loc[indices]
print(df_pairwise.shape)
df_pairwise.head()
```

```python
df_contratos.n_contracts.sort_values(ascending=False)
```

```python
df_red = (df_pairwise.groupby('par', as_index=False).tender_id.count()
                     .rename(columns={'tender_id': 'weigth'}))
df_red = df_red.join(df_red.par.str.split('|', expand=True)).drop('par', axis=1)
df_red = df_red.rename(columns={0: 'source', 1: 'target'})
# weight min
# W = 2
# df_red = df_red.loc[df_red.weigth >= W]

print(df_red.shape)
df_red.head()
```

```python
all_nodes = list(set(df_red.source.unique()) | set(df_red.target.unique()))
df_all_nodes = pd.DataFrame(all_nodes, columns=['supplier_id'])
print(df_all_nodes.shape)
df_all_nodes.head(2)
```

```python
df_contracts_e = df_contracts.loc[(df_contracts.entidad_id == e)]
df_contracts_e = df_contracts_e.loc[:, ['tender_id', 'supplier_id', 'contract_amount']].drop_duplicates()
# monto
df_monto = (df_contracts_e.groupby('supplier_id', as_index=False).contract_amount.sum()
                          .rename(columns={'contract_amount': 'contract_amount'}))
# contratos
df_contratos = (df_contracts_e.groupby('supplier_id', as_index=False).contract_amount.count()
                              .rename(columns={'contract_amount': 'n_contracts'}))

df_nodos = pd.merge(df_monto, df_contratos, on='supplier_id', how='inner')
df_nodos = pd.merge(df_all_nodes, df_nodos, on='supplier_id', how='left')
df_nodos = df_nodos.assign(
    contract_amount=df_nodos.contract_amount.fillna(0),
    n_contracts=df_nodos.n_contracts.fillna(0),
)
df_nodos.head()
```

```python
df_nodos = df_nodos.assign(
    # monto_por_contrato=df_nodos.contract_amount.divide(df_nodos.n_contracts),
    # size_log=(np.log(df_nodos.contract_amount + 10) * 5)
    size_log=((df_nodos.n_contracts + 1) * 2)
)
# df_nodos = df_nodos.assign(contract_amount_log=np.log(df_nodos.contract_amount + 2)) * 3
print(df_nodos.shape)
df_nodos.head(2)
```

```python
# raro
# e = 'CONAGUA-94'
# u = 'CNA890116SF2-016B00985'

# normal
e = 'SCT-262'
u = 'SCT051121IB4-009000967'

df_raros = (df_participantes.loc[(df_participantes.entidad_id == e) & (df_participantes.unidad_id == u)] 
                            .groupby(['tender_id', 'supplier_id']).tender_status.count()
                            .reset_index()
                            .rename(columns={'tender_status': 'valor'})
                            .drop('valor', axis=1))

df_contracts_aux = df_contracts.loc[(df_contracts.entidad_id == e) & (df_contracts.unidad_id == u)]
df_contracts_aux = df_contracts_aux.loc[:, ['tender_id', 'supplier_id', 'contract_amount']].drop_duplicates()
df_raros = pd.merge(
    df_raros, df_contracts_aux,
    on=['tender_id', 'supplier_id'], how='left'
)
print(df_raros.shape)
df_raros = df_raros.pivot(index='supplier_id', columns='tender_id', values='contract_amount')
suppliers_orden = df_raros.sum(axis=1).sort_values(ascending=False).index
tender_orden = (df_contracts.loc[(df_contracts.entidad_id == e) & (df_contracts.unidad_id == u), 
                                 ['tender_id', 'contract_period_start']]
                            .drop_duplicates()
                            .sort_values('contract_period_start').tender_id.values)

df_raros = df_raros.loc[suppliers_orden, tender_orden]
df_raros
```

```python
# Normal?
# e = 'SCT-262'
# u = 'SCT051121IB4-009000967'
with plt.style.context('seaborn-talk'):
    fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(25, 22))
    sns.heatmap(df_raros, cmap='viridis', ax=ax)
```
